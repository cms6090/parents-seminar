## 네이버 블로그 크롤링

```python
# 라이브러리 불러오기
from selenium import webdriver 
import requests
from bs4 import BeautifulSoup as BS
import pandas as pd
import time
import re
import pickle
from selenium.webdriver.common.by import By
import datetime
```

```python
# 크롬 드라이버 로드
file_path = "C:\\Users\\cho03\\Downloads\\세미나\\방학 세미나\\"
path = file_path + 'chromedriver.exe'
driver = webdriver.Chrome(path)
```

```python
# 네이버 블로그 검색 페이지로 이동
driver.get('https://search.naver.com/search.naver?where=blog&query=%EB%85%B8%EC%9D%B8+%EB%B6%80%EC%96%91&sm=tab_opt&nso_open=1&nso=so:r,p:from20180705to20230705,a:all')
```

```python
# 스코롤 다운 함수 정의
def doScrollDown(whileSeconds): 
    start = datetime.datetime.now()
    end = start + datetime.timedelta(seconds=whileSeconds)
    while True:
        driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        time.sleep(1)
        if datetime.datetime.now() > end:
            break
```

```python
# 페이지 스크롤 다운
doScrollDown(35)
```

```python
# 제목과 URL을 저장할 리스트 초기화
title_list = []
url_list = []

# 블로그 글 목록을 크롤링
for i in range(1,1100):
    try:
        title = driver.find_element(By.XPATH,'//*[@id="sp_blog_{}"]/div/div/a'.format(i)).text # 제목
        title2 = driver.find_element(By.XPATH, '//*[@id="sp_blog_{}"]/div/div/a'.format(i))    # URL
        url = title2.get_attribute('href')
    
        title_list.append(title) #제목 추가
        url_list.append(url) #url주소 추가
        if (i % 10) == 0:
            print(i)
    except:
        break
```

```python
# 본문, 좋아요 수, 댓글 수, 댓글, 이미지 수, 영상 수를 저장할 리스트 초기화
new_doc = []
like_cnt = []
comment_cnt = []
comment_list = []
img_cnt = []
div_cnt = []

# 블로그 글 크롤링
for i in range(len(url_list)):
    url_path = url_list[i]
    driver.switch_to.window(driver.window_handles[0])
    driver.execute_script("window.open('{}')".format(url_path))
    driver.switch_to.window(driver.window_handles[1])
    
    time.sleep(1)
    try:
        iframes = driver.find_elements(By.TAG_NAME, 'iframe')
        d = ''

        if len(iframes) > 0:
            driver.switch_to.frame(0)
            html = driver.page_source
            soup = BS(html, 'html.parser')

            try:
                a = soup.find('div', class_= 'se-main-container').get_text()  

            except:
                a = soup.find('div', id='postListBody')
                a = re.sub("[^ㄱ-ㅎㅏ-ㅣ가-힣]", "", str(a))

            try:
                b = soup.find('em', class_='u_cnt _count').get_text()
            except:
                b = 'null'

            try:
                c = soup.find('em', id="commentCount").get_text()
            except:
                c = 'null'

            try:
                comment = driver.find_elements(By.CLASS_NAME, 'btn_arr')
                comment[len(comment) - 1].click()
                time.sleep(1)

                commentLen = len(driver.find_elements(By.CLASS_NAME, 'u_cbox_page'))
                d = ''
                for _ in range(0, commentLen) :
                    commentList = driver.find_elements(By.CLASS_NAME, 'u_cbox_text_wrap')
                    for j in range(len(commentList)):
                        commentList[j] = commentList[j].text
                soup = BS(html, 'html.parser')
                d = "\n".join(commentList)

            except:
                d = 'null'

            e = len(soup.find_all('img', class_='se-image-resource egjs-visible'))
            f = len(soup.find_all('div', class_='pzp-ui-dimmed pzp-dimmed pzp-pc__dimmed'))

            new_doc.append(a)
            like_cnt.append(b)
            comment_cnt.append(c)
            comment_list.append(d)
            img_cnt.append(e)
            div_cnt.append(f)
            driver.switch_to.default_content()

        else:
            new_doc.append(' ')
            like_cnt.append(' ')
            comment_cnt.append(' ')
            comment_list.append(' ')
            img_cnt.append(' ')
            div_cnt.append(' ')
            
    except:
        new_doc.append(' ')
        like_cnt.append(' ')
        comment_cnt.append(' ')
        comment_list.append(' ')
        img_cnt.append(' ')
        div_cnt.append(' ')
    
    driver.close()
    time.sleep(0.3)
    if(i%10) == 0:
        print(i)
```

```python
# 크롤링한 데이터를 데이터프레임으로 변환
raw_data = pd.DataFrame()
raw_data['title'] = title_list
raw_data['doc'] = new_doc
raw_data['like'] = like_cnt
raw_data['comment_cnt'] = comment_cnt
raw_data['comment_list'] = comment_list
raw_data['img'] = img_cnt
raw_data['div'] = div_cnt
raw_data['ch'] = 'naver'
raw_data['ch2'] = 'blog'
```

```python
# 데이터프레임을 pickle파일로 저장
f = open(file_path + "naver_blog.pkl", "wb")
pickle.dump(raw_data,f)

f.close()
```

```python
# 크롬 드라이버 종료
driver.quit()
```
