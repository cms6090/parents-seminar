# 긍부정 상관분석

```python
import pickle
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
file_path = 'C:\\Users\\cho03\\Downloads\\세미나\\방학 세미나\\' # data가 저장된 파일 경로
```

```python
# 불러오기
f = open(file_path + "total_docs1.pkl", "rb")
data1 = pickle.load(f)
f.close()
```

```python
ff = data1[['like','img','div','comment_cnt','values']]

data = pd.DataFrame()
data['like'] = ff['like']
data['img'] = ff['img']
data['div'] = ff['div']
data['comment_cnt'] = ff['comment_cnt']
data['values'] = ff['values']
data
```

```python
# 'values' 열에서 -2부터 2까지의 각 값들의 개수를 카운트하여 새로운 열로 추가
data['매우 부정'] = data['values'].apply(lambda x: x.count(-2))
data['부정'] = data['values'].apply(lambda x: x.count(-1))
data['중립'] = data['values'].apply(lambda x: x.count(0))
data['긍정'] = data['values'].apply(lambda x: x.count(1))
data['매우 긍정'] = data['values'].apply(lambda x: x.count(2))

# 'values' 열 삭제
data.drop('values', axis=1, inplace=True)

# 결과 출력
data
```

```python
df = data[data['comment_cnt'] != '<NA>']
df = df[df['like'] != '<NA>']

df = df.reset_index(drop=True)
df
```

```python
corr = df.corr(method='pearson')
corr
```

```python
plt.rc('font', family = 'Malgun Gothic')

# 상관계수 행렬 시각화
plt.figure(figsize=(15, 15))
sns.heatmap(corr, annot=True, cmap='Blues', fmt=".2f", linewidths=0.5)
```

```python
var1 = []
var2 = []
pvalue = []
for i in range(14):
    for j in range(i+1,9):
        if (i == j):
            continue
        rho, p_val = stats.pearsonr(df[df.columns[i]],df[df.columns[j]])
        var1.append(df.columns[i])
        var2.append(df.columns[j])
        pvalue.append(p_val)
```

```python
pd.set_option('display.max_rows', 100)
```

```python
c = pd.DataFrame()
c['변수 1'] = var1
c['변수 2'] = var2
c['p-value'] = pvalue
c
```

# 10가지 감성 상관분석

```python
import pickle

# 불러오기
f = open(file_path + "total_docs2.pkl", "rb")
ff = pickle.load(f)
f.close()
```

```python
data1 = pd.DataFrame()
data1['like'] = ff['like']
data1['img'] = ff['img']
data1['div'] = ff['div']
data1['comment'] = ff['comment_cnt']
data1['values'] = ff['values']
data1
```

```python
data = data1[data1['values'].apply(lambda x: len(x) == 10)]
data['anger'] = data['values'].apply(lambda x: x[0])
data['anticipation'] = data['values'].apply(lambda x: x[1])
data['disgust'] = data['values'].apply(lambda x: x[2])
data['fear'] = data['values'].apply(lambda x: x[3])
data['joy'] = data['values'].apply(lambda x: x[4])
data['negative'] = data['values'].apply(lambda x: x[5])
data['positive'] = data['values'].apply(lambda x: x[6])
data['sadness'] = data['values'].apply(lambda x: x[7])
data['surprise'] = data['values'].apply(lambda x: x[8])
data['trust'] = data['values'].apply(lambda x: x[9])
data.drop('values', axis=1, inplace=True)

data
```

```python
df = data[data['comment'] != '<NA>']
df = df[df['like'] != '<NA>']

df = df.reset_index(drop=True)
df
```

```python
corr = df.corr(method='pearson')
corr
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(corr, annot=True, linewidths=.5, cmap='Blues')
```

```python
var1 = []
var2 = []
pvalue = []
for i in range(14):
    for j in range(i+1,14):
        if (i == j):
            continue
        rho, p_val = stats.pearsonr(df[df.columns[i]],df[df.columns[j]])
        var1.append(df.columns[i])
        var2.append(df.columns[j])
        pvalue.append(p_val)
```

```python
c = pd.DataFrame()
c['변수 1'] = var1
c['변수 2'] = var2
c['p-value'] = pvalue
c
```

```python
pd.set_option('display.max_rows', 10)
```
