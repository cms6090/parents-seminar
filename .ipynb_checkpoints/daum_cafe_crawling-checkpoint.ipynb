```python
from selenium import webdriver
import requests
from bs4 import BeautifulSoup as BS
import pandas as pd
import time
import re
import pickle
from selenium.webdriver.common.by import By
```

## 다음 카페 크롤링

```python
query_txt = "노부모 갈등"
```

```python
file_path = "C:\\Users\\cho03\\Downloads\\세미나\\방학 세미나\\"
path = file_path + 'chromedriver.exe'
driver = webdriver.Chrome(path)
```

```python
driver.get('https://search.daum.net/search?w=fusion&col=cafe&q=%EB%85%B8%EB%B6%80%EB%AA%A8+%EA%B0%88%EB%93%B1&DA=STC&sd=20180705000000&ed=20230705235959&period=u')
```

```python
title_list = []
url_list = []

for j in range(0,10):
    for i in range(1,16):
        try:
            title = driver.find_element(By.XPATH,'//*[@id="twcColl"]/div[2]/c-container/c-card[{}]/div/c-doc-web/div/div[1]'.format(i)).text
            title2 = driver.find_element(By.XPATH, '//*[@id="twcColl"]/div[2]/c-container/c-card[{}]/div/c-doc-web/div/div[1]/c-title'.format(i))
            url = title2.get_attribute('data-href')
    
            title_list.append(title)
            url_list.append(url)
        except:
            break
    driver.find_element(By.CLASS_NAME, 'btn_next').click()
```

```python
new_doc = []
like_cnt = []
comment_cnt = []
comment_list = []
img_cnt = []
div_cnt = []

for i in range(len(url_list)):
    url_path = url_list[i]
    driver.switch_to.window(driver.window_handles[0])
    driver.execute_script("window.open('{}')".format(url_path))
    driver.switch_to.window(driver.window_handles[1])
    time.sleep(1)
    iframes = driver.find_elements(By.TAG_NAME, 'iframe')
    if len(iframes) > 0:
        driver.switch_to.frame(0)
        html = driver.page_source
        soup = BS(html, 'html.parser')
        try:
            a = soup.find('table', class_= 'protectTable').get_text().strip()
            like = soup.find('button', {'id': 'recommend-btn'})
            b = int(like.find('span', class_='txt_num').get_text())
            comment = soup.find('button', {'id':'comment-btn'})
            c = int(comment.find('span', class_='txt_num').get_text())
            commentlist = soup.find_all('span', class_='original_comment')
            d = "\n".join(item.get_text() for item in commentlist)
            img = soup.find_all('img', class_='txc-image')
            e = len(img)
            div = soup.find_all('div', class_='figure-video')
            f = len(div)
                
        except:
            a = 'null'
        new_doc.append(a)
        like_cnt.append(b)
        comment_cnt.append(c)
        comment_list.append(d)
        img_cnt.append(e)
        div_cnt.append(f)
        driver.switch_to.default_content()
    
    else:
        new_doc.append(' ')
        like_cnt.append(' ')
        comment_cnt.append(' ')
        comment_list.append(' ')
        img_cnt.append(' ')
        div_cnt.append(' ')
        
    driver.close()
    time.sleep(2)
    print(i+1)
```

```python
raw_data = pd.DataFrame()
raw_data['title'] = title_list
raw_data['doc'] = new_doc
raw_data['like'] = like_cnt
raw_data['comment_cnt'] = comment_cnt
raw_data['comment_list'] = comment_list
raw_data['img'] = img_cnt
raw_data['div'] = div_cnt
raw_data['ch'] = 'daum'
raw_data['ch2'] = 'cafe'
```

```python
f = open(file_path + "daum_cafe.pkl", "wb")
pickle.dump(raw_data,f)

f.close()
```
