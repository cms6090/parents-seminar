## 텍스트 데이터 마이닝

```python
import pickle
import pandas as pd
```

```python
file_path = "C:\\Users\\user\\Documents\\랩실\\2-1\\서교수님 크롤링\\방학 세미나_최종\\" # 파일 경로
```

```python
f = open(file_path + 'total_doc.pkl', "rb") # 데이터 불러오기
data = pickle.load(f)
f.close()
data
```

### 제목 데이터 마이닝

```python
import itertools # 문서의 명사 리스트 확보
noun = list(itertools.chain(*data['title_token_noun'])) # 리스트 접합
noun
```

```python
from collections import Counter
count = Counter(noun) # 각 문자가 몇 번 나타나는지 알려주는 객체 반환
```

```python
top = dict(count.most_common(100)) # 상위 100개 출력
top
```

```python
# csv 파일로 저장
import csv

with open("제목 top100.csv", 'w', newline = '') as file:
    writer = csv.writer(file)
    for k, v in top.items():
        writer.writerow([k, v])
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# 글씨체 위치 지정
wordcloud = WordCloud(font_path = file_path + 'NanumGothic\\NanumGothicBold.ttf', background_color = 'white', colormap = "Accent", width = 600, height = 400).generate_from_frequencies(top)
plt.figure(figsize=(12,8))   # 보여지는 크기 지정
plt.imshow(wordcloud) # 이미지 표시
plt.axis('off') #눈금 제거
plt.show()
```

### 전체 데이터 마이닝

```python
import itertools # 문서의 명사 리스트 확보
noun = list(itertools.chain(*data['title_token_noun'])) # 리스트 접합
noun += list(itertools.chain(*data['doc_token_noun']))
noun += list(itertools.chain(*data['comment_token_noun']))
noun
```

```python
from collections import Counter
count = Counter(noun) # 각 문자가 몇 번 나타나는지 알려주는 객체 반환
```

```python
top = dict(count.most_common(100)) # 상위 100개 출력
top
```

```python
# csv 파일로 저장
import csv

with open("전체 데이터 top100.csv", 'w', newline = '') as file:
    writer = csv.writer(file)
    for k, v in top.items():
        writer.writerow([k, v])
```
